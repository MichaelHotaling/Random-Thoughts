```python
from langchain.agents import create_pandas_dataframe_agent
from langchain.agents.agent_types import AgentType
from langchain.chat_models import ChatOpenAI
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
agent = create_pandas_dataframe_agent(ChatOpenAI(temperature=0, model='gpt-4'), df, verbose=True)
```

```python
agent.run("how many rows are there?")
```

```python
agent.run("Create a decision tree model with the target variable being survived or not. \
Split the data into 50/50 training and testing sets, build the model using the training set and \
test the model with the test set. plot the ROC curve of the test set. Your final action should be to \
create a ROC curve and save it as a PNG file.")
```
